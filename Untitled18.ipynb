```python
# prompt: funcio x al cuadrado graficarla

import matplotlib.pyplot as plt
import numpy as np

def x_squared(x):
  return x**2

x = np.linspace(-10, 10, 400)
y = x_squared(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("x^2")
plt.title("Función x al cuadrado")
plt.grid(True)
plt.show()
```
